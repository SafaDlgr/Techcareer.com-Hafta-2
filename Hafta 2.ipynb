```


###  Soru 1 – Sayı Analizi
# Kullanıcıdan bir sayı isteyin.
#• Sayı pozitif, negatif ya da sıfır mı kontrol edin.
#• Aynı zamanda tek/çift olup olmadığını da belirtin.
#• Çıktı örneği: "Pozitif Çift" veya "Negatif Tek" gibi.

"""

sayi = float(input("Lütfen bir sayı giriniz: "))

if sayi == 0:
    print("Sayı sıfırdır.")
elif sayi > 0:
    if sayi % 2 == 0:
        print("Sayı pozitif çift sayıdır.")
    else:
        print("Sayı pozitif tek sayıdır.")
else:
    if sayi % 2 == 0:
        print("Sayı negatif çift sayıdır.")
    else:
        print("Sayı negatif tek sayıdır.")

"""

### Soru 2 – Harf Frekansı (String)
# Kullanıcıdan bir kelime alın.
#• Hangi harften kaç tane geçtiğini bulun.
#• Sonucu dictionary olarak gösterin.
#Örnek: "data" → {'d': 1, 'a': 2, 't': 1}

"""

kelime = input("Lütfen bir kelme giriniz: ")
liste = {}
for harf in kelime:
    if harf in liste:
        liste[harf] += 1
    else:
        liste[harf] = 1
print(liste)

"""


### Soru 3 – Şifre Kontrolü (String Metotları)
# Kullanıcıdan şifre girmesini isteyin. Şifre:
#• En az 8 karakter olmalı
#• En az 1 büyük harf içermeli
#• En az 1 rakam içermeli
#Koşulları sağlayıp sağlamadığına göre kullanıcıyı bilgilendirin.

"""

print("Şifreniz en az 8 karakter olmalı, en az bir büyük harf ve sayı içermelidir.")
sifre = input("Lütfen şifrenizi belirleyiniz: ")
uzn = len(sifre)
buyuk = False
rakam = False

for karakter in sifre:
    if karakter.isupper():
        buyuk = True
    if karakter.isdigit():
        rakam = True

if uzn >= 8 and rakam and buyuk:
    print("Şifre geçerli.")
else:
    if uzn < 8:
        print("Şifre ne az 8 karakter olmalıdır.")
    if not buyuk:
        print("Şifre en az 1 büyük harf içermelidir.")
    if not rakam:
        print("Şifre en az 1 rakam içermelidir.")

"""


### Soru 4 – Liste İşlemleri
# [12, 4, 9, 25, 30, 7, 18] listesini kullanın.
#• Listenin ortalamasını bulun.
#• Ortalamadan büyük sayıları ayrı bir listeye atın.
#• Sonucu ekrana yazdırın.

"""

liste = [12, 4, 9, 25, 30, 7, 18]
toplam = 0

for i in liste:
    toplam += i

ortalama = toplam / len(liste)

buyukler = []

for sayi in liste:
    if sayi > ortalama:
        buyukler.append(sayi)

print("Toplam:", toplam)
print("Ortalam:", ortalama)
print("Ortalamadan büyük sayıalar:", buyukler)    

"""
    
### Soru 5 – Nested Loop (Desen)
#Aşağıdaki çıktıyı üreten programı yazın (üçgen desen):
# *
# **
# ***
# ****
# *****

"""

for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

"""



### Soru 6 – While Döngüsü
# Kullanıcıdan sürekli sayı isteyin.
#• Kullanıcı 0 girdiğinde program dursun.
#• Girilen tüm sayıların toplamını ve ortalamasını yazdırın.

"""

toplam = 0 
adet = 0
print("Çıkış yapmak için 0 giriniz.")
while True:
    sayi = float(input("Lütfen bir sayı giriniz: "))
    if sayi == 0:
        break
    toplam += sayi
    adet += 1
if adet > 0:
    ortalama = toplam / adet
    print("Ortalama:", ortalama)
    print("Toplam:", toplam)
else:
    print("Herhangi bir sayı girilmedi.")

"""


### Soru 7 – Palindrom Kontrolü
# Kullanıcıdan bir kelime isteyin.
#• Kelimenin palindrom olup olmadığını kontrol edin.
#• Örnek: "kayak" → Palindrom , "python" → Değil 

"""

kelime = input("Bir kelime giriniz: ")
# kelime = kelime.lower() büyük küçük harf önemli değilse kullanılabilir.

if kelime == kelime[::-1]:
    print("Kelime palindromdur.")
else:
    print("Kelime palindrom değildir.")

"""

###Soru 8 – List Comprehension
# 1’den 100’e kadar olan sayılardan:
#• Hem 3’e hem 5’e bölünebilenlerin karelerini içeren bir liste oluşturun.
#• Sonucu ekrana yazdırın.

"""

liste = [i**2 for i in range(1, 101) if i % 3 == 0 and i % 5 == 0]
print(liste)

"""

### Soru 9 – String İşlemleri
# Bir cümle alın.
#• Cümledeki kelimeleri split() ile ayırın.
#• Her kelimenin ilk harfini büyük yaparak yeni bir string oluşturun.
#• Örnek: "python veri bilimi" → "Python Veri Bilimi"

"""

cumle = input("Bir cümle girin: ")
kelimeler_ilk = cumle.split()

kelimeler_son = []
for kelime in kelimeler_ilk:
    kelimeler_son.append(kelime.capitalize())

sonuc = " ".join(kelimeler_son)
print(sonuc)

"""

### Mini Proje – Film Yorumu Analizi
# Proje Tanımı:
# Kullanıcıdan birkaç film yorumu alın (örneğin 5–6 yorum).
#- Yorumları bir listeye atın.
#- Her yorumun uzunluğunu (karakter sayısı) bulun.
#- Kaç yorumda "iyi" kelimesi geçtiğini sayın.
#- En uzun yorumu ve en kısa yorumu ekrana yazdırın.
#- Tüm yorumların ortalama uzunluğunu hesaplayın.
# :Beklenen Çıktı Örneği
# Girdi (yorumlar):
# ["Film çok güzeldi", "Kötüydü", "Ortalama bir filmdi", "Gerçekten çok iyi!", "İyi ama daha iyi olabilirdi"]
# Çıktı:
# Toplam yorum sayısı: 5
# "iyi" geçen yorum sayısı: 2
# En uzun yorum: İyi ama daha iyi olabilirdi
# En kısa yorum: Kötüydü
# Ortalama uzunluk: 17.4 karakter

"""

yorumlar = []
for i in range(5):
    yorum = input("Yorum girin: ")
    yorumlar.append(yorum)

uzunluklar = []
for i in yorumlar:
    uzunluklar.append(len(i))

iyi = 0
for yorum in yorumlar:
    if "iyi" in yorum.lower():
        iyi += 1

uzun = max(yorumlar, key=len)
kısa = min(yorumlar, key=len)

toplam_uzunluk = sum(uzunluklar)
ortalama = toplam_uzunluk / len(yorumlar)

print("Yorumlar:", yorumlar)
print("Toplam yorum sayısı:", len(yorumlar))
print('"iyi" geçen yorum sayısı:', iyi)
print("En uzun yorum:", uzun)
print("En kısa yorum:", kısa)
print("Ortalama uzunluk:", ortalama, "karakter")

"""



```
